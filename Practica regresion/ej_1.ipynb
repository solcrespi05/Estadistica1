```python
import numpy as np
import matplotlib.pyplot as plt
"""
1. Consideremos la funci√≥n ùëì : R ‚Üí R dada por ùëì (ùë•) = 2ùë• ‚àí1.
(a) Seanùëã ‚àº U(0,1) yùëå := ùëì (ùëã). Fijemosùëõ = 100. Tomar muestrasùë•1,...,ùë•ùëõ deùëã y aplicar la
funci√≥n ùëì a cada muestra para calcularùë¶ùëñ = ùëì (ùë•ùëñ), con 1 ‚â§ ùëñ ‚â§ ùëõ. Realizar un ajuste lineal de
la forma ùëå = ùëãùõΩ1 + ùõΩ0 +ùúñ a partir de los samples (ùë•1,ùë¶1),..., (ùë•ùëõ,ùë¶ùëõ) generados y calcular,
para cada ùëñ ‚àà {1,...,ùëõ}, ùúÄùëñ = ùë¶ùëñ ‚àí (ùë•ùëñùõΩ1 + ùõΩ0).
 Describir la distribuci√≥n emp√≠rica dada por {ùúÄ1, . . .,ùúÄùëõ}.
 """

#El enunciado pide 100 muestras de una distribucion uniforme en (0,1)
muestras_x = np.random.uniform(0,1,100)

#La funcion f(x) = 2x - 1
muestras_y = 2*muestras_x - 1

#Usamos minimos cuadrados para ajustar los parametros B1 y B0
#Ac√° tenemos dos formas de hacerlo, una es con la formula de minimos cuadrados y la otra es con np.polyfit
#La formula de minimos cuadrados es:
B1 = np.sum((muestras_x - np.mean(muestras_x))*(muestras_y - np.mean(muestras_y)))/np.sum((muestras_x - np.mean(muestras_x))**2)
B0 = np.mean(muestras_y) - B1*np.mean(muestras_x)
print('B1:', B1)
print('B0:', B0)

#La funcion np.polyfit nos devuelve los coeficientes de la regresion polinomica
beta1,beta0 = np.polyfit(muestras_x, muestras_y, 1)

#calculamos los residuales
residuales = muestras_y - (beta1*muestras_x + beta0)


# Calcular la media y la varianza de los residuales
media_residuales = np.mean(residuales)
varianza_residuales = np.var(residuales)
print('Media de los residuales:', media_residuales)
print('Varianza de los residuales:', varianza_residuales)

# Graficar el histograma de los residuales
plt.hist(residuales, bins=20, edgecolor='black')
plt.title('Distribuci√≥n emp√≠rica de los residuales')
plt.show()
```

```python
'''Sean ùëã ‚àº U(0,1),ùëç ‚àº N(0,0.25) yùëå := ùëì (ùëã)+ùëç. Fijemos ùëõ = 100. Tomar muestras ùë•1,...,ùë•ùëõ
 de ùëã y ùëß1,...,ùëßùëõ de ùëç y calcular ùë¶ùëñ = ùëì (ùë•ùëñ) +ùëßùëñ, con 1 ‚â§ ùëñ ‚â§ ùëõ. Realizar un ajuste lineal de
 la forma ùëå = ùëãùõΩ1 + ùõΩ0 +ùúñ a partir de los samples (ùë•1,ùë¶1),..., (ùë•ùëõ,ùë¶ùëõ) generados y calcular,
 para cada ùëñ ‚àà {1,...,ùëõ}, ùúÄùëñ = ùë¶ùëñ ‚àí (ùë•ùëñùõΩ1 + ùõΩ0).
 Describir la distribuci√≥n emp√≠rica dada por {ùúÄ1, . . .,ùúÄùëõ} haciendo un histograma.
 '''

muestras_x = np.random.uniform(0,1,100)
muestras_z = np.random.normal(0,0.25,100)

muestras_y = 2*muestras_x - 1 + muestras_z

b1 = np.sum((muestras_x - np.mean(muestras_x))*(muestras_y - np.mean(muestras_y)))/np.sum((muestras_x - np.mean(muestras_x))**2)
b0 = np.mean(muestras_y) - b1 * np.mean(muestras_x)

residuales = muestras_y - (muestras_x*b1 + b0)

# Graficar el histograma de los residuales
plt.hist(residuales, bins=20, edgecolor='black')
plt.title('Distribuci√≥n emp√≠rica de los residuales')
plt.show()
```
